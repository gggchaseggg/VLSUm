Два аудитора проверяют N фирм (равное число каждый), у М из которых имеются нарушения. Вероятность обнаружения нарушений первым  аудитором равна p1, вторым — р2. Найти вероятность того, что все фирмы-нарушители будут выявлены, если:
- a) N = 6, М = 2, p1 = 0,7, р2 = 0,8;
- б) N = 6, М = 3, p1 = 0,7, р2 = 0 9;
- в) N = 8, М = 3, p1 = 0,6, р2 = 0,8;
- г) N = 10, М = 2, p1 = 0,8, р2 = 0,9.

```python
import matplotlib.pyplot as plt
import numpy as np

def derevo(N, M,p1, p2):
  P_detect = 1 - (1 - p1) * (1 - p2)
  P_all_detected = P_detect ** M

  print(f"Вероятность того, что все {M} фирмы-нарушители будут выявлены: {P_all_detected:.4f}")

  fig, ax = plt.subplots(figsize=(10, 6))

  ax.text(0.1, 0.9, 'Все фирмы', fontsize=14, ha='center')
  ax.text(0.3, 0.7, f'1-й аудитор выявил\n({p1 * 100:.0f}%)', fontsize=12, ha='center')
  ax.text(0.3, 0.5, f'1-й аудитор не выявил\n({(1 - p1) * 100:.0f}%)', fontsize=12, ha='center')
  ax.text(0.6, 0.75, f'2-й аудитор выявил\n({p2 * 100:.0f}%)', fontsize=10, ha='center')
  ax.text(0.6, 0.65, f'2-й аудитор не выявил\n({(1 - p2) * 100:.0f}%)', fontsize=10, ha='center')
  ax.text(0.6, 0.55, f'2-й аудитор выявил\n({p2 * 100:.0f}%)', fontsize=10, ha='center')
  ax.text(0.6, 0.45, f'2-й аудитор не выявил\n({(1 - p2) * 100:.0f}%)', fontsize=10, ha='center')
  ax.plot([0.1, 0.3], [0.9, 0.7], 'k-')
  ax.plot([0.1, 0.3], [0.9, 0.5], 'k-')
  ax.plot([0.3, 0.6], [0.7, 0.75], 'k-')
  ax.plot([0.3, 0.6], [0.7, 0.65], 'k-')
  ax.plot([0.3, 0.6], [0.5, 0.55], 'k-')
  ax.plot([0.3, 0.6], [0.5, 0.45], 'k-')
  ax.axis('off')
  plt.show()

derevo(6, 2, 0.7, 0.8)
derevo(6, 3, 0.7, 0.9)
derevo(8, 3, 0.6, 0.8)
derevo(10, 2, 0.8, 0.9)
```

В урне К белых шаров, L красных и М черных. Найти вероятность достать (без возвращения) три шара, не все цвета которых одинаковы, если
- а) К = 9, L = 7, М = 4;
- б) К = 10, L = 8, М = 6.

```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt

def probability_diff_colors(K, L, M):
    total_combinations = comb(K + L + M, 3)
    white_combinations = comb(K, 3) if K >= 3 else 0
    red_combinations = comb(L, 3) if L >= 3 else 0
    black_combinations = comb(M, 3) if M >= 3 else 0
    one_color_combinations = white_combinations + red_combinations + black_combinations
    prob_one_color = one_color_combinations / total_combinations
    prob_diff_colors = 1 - prob_one_color

    return prob_diff_colors

params = [(9, 7, 4), (10, 8, 6)]
probabilities = [probability_diff_colors(K, L, M) for K, L, M in params]

for i, (K, L, M) in enumerate(params):
    print(f"Случай {chr(97 + i)}: K = {K}, L = {L}, M = {M}, вероятность разного цвета = {probabilities[i]:.4f}")

K_vals = np.arange(3, 15)
L_vals = [5, 6, 7, 8]
colors = ['blue', 'green', 'orange', 'red']
plt.figure(figsize=(10, 6))

for i, L in enumerate(L_vals):
    probabilities = [probability_diff_colors(K, L, 4) for K in K_vals]
    plt.plot(K_vals, probabilities, label=f'L = {L}', color=colors[i])

plt.xlabel('Количество белых шаров (K)')
plt.ylabel('Вероятность хотя бы одного другого цвета')
plt.title('Вероятность разного цвета шаров при различных K и L')
plt.legend()
plt.grid(True)
plt.show()
```
