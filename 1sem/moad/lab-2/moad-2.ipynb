# Задание 1
Среди 5 ключей два подходят к двери. Ключи пробуют один за другим, пока не откроют дверь. Найти распределение числа опробованных ключей.

```python
total_keys = 5
fitting_keys = 2

p_first_try = fitting_keys / total_keys
p_second_try = (total_keys - 1) / total_keys * fitting_keys / (total_keys - 1)
p_third_try = (total_keys - 2) / total_keys * (total_keys - 1) / (total_keys - 1) * fitting_keys / (total_keys - 2)

print(f'Вероятность открыть дверь с первого ключа: {p_first_try:.2f}')
print(f'Вероятность открыть дверь со второго ключа: {p_second_try:.2f}')
print(f'Вероятность открыть дверь с третьего ключа: {p_third_try:.2f}')
```

```python
import math

C = 1 / (9 * math.pi)

def F_x(x):
    if -3 <= x <= 2:
        return C * (9 - x ** 2)
    else:
        return 0

M_xi = C * sum([9 - x ** 2 for x in range(-3, 3)])
P_A = C * sum([9 - x ** 2 for x in range(0, 3)])

print(f'Константа C: {C:.4f}')
print(f'Математическое ожидание: {M_xi:.4f}')
print(f'Вероятность P(A): {P_A:.4f}')
```

# Задание 3
В среднем 10% работоспособного населения региона — безработные. Оценить с помощью неравенства Чебышева вероятность того, что уровень безработицы среди случайно выбранных 10 000 жителей составит от 9 до 11%.

```python
p = 0.1
n = 10000

mean = p * n
variance = n * p * (1 - p)

low = 0.09 * n
high = 0.11 * n

probability = 1 - (variance / ((high - mean) ** 2))

print(f'Вероятность того, что уровень безработицы будет в пределах 9-11%: {probability:.4f}')
```

# Задание 4
Случайная величина ξ равномерно распределена на отрезке L. Найти плотность распределения случайной величины η, если: I = [1, 2]; η = 6/(ξ+1).

```python
I = [1, 2]

def density_eta(x):
    if I[0] <= x <= I[1]:
        return 6 / (x + 1)
    else:
        return 0

eta_values = [density_eta(x) for x in range(I[0], I[1] + 1)]

print(f'Значения плотности распределения η: {eta_values}')
```
